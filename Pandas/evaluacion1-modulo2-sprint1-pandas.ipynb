3. PreparaciÃ³n de los datos:

- Los nombres de las columnas empiezan con mayÃºsculas, poned todos los nombres de las columnas en minÃºsculas.

```python
'''Homogeneizamos columnas'''

nuevas_columnas = {col: col.lower() for col in df_spoti.columns}

df_spoti.rename(columns = nuevas_columnas, inplace = True)
```

- Haciendo el anÃ¡lisis exploratorio nos deberÃ­amos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberÃ­an. Esto es debido a que los decimales estÃ¡n establecidos como comas y no con puntos. Cread una funciÃ³n que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.

- DespuÃ©s de haber hecho los cambios, chequead los tipos de datos. Â¿Son ya del tipo correcto? En caso de que no, cread otra funciÃ³n o modificar la anterior para que sean de tipo float.

```python
'''Como quise cambiar todas las columnas a la vez y eso requiere mÃ¡s de un argumentos, busquÃ© una soluciÃ³n con una lambda para poder usar el apply'''

def reemplazar(df, columnas):
    df[columnas] = df[columnas].apply(lambda x: x.str.replace(',', '.')).astype(float)
    return df
```

```python
reemplazar(df_spoti, ['danceability','energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness'])

df_spoti.head(2)
```

- Hay algunas canciones cuyo tÃ­tulo estÃ¡ en mayÃºsculas. Crea una funciÃ³n para que todos los tÃ­tulos estÃ©n en minÃºscula. Haced lo mismo para las columnas de "track" y "album".

```python
df_spoti['title'].unique()
```

```python
'''Creamos una funciÃ³n para cambiar los titulos a minÃºsculas, y le damos la comdiciÃ³n de que compruebe si el tipo del tÃ­tulo
es tipo string, para que en ese caso le aplique un .lower()'''

def minusculas(titulo):
    if isinstance(titulo, str):
        return titulo.lower()
    else:
        return titulo
```

```python
'''Usamos la funciÃ³n sobre todas las columnas que queremos hacer el cambio con el apply y la funciÃ³n'''

df_spoti['title'] = df_spoti['title'].apply(minusculas)
df_spoti['track'] = df_spoti['track'].apply(minusculas)
df_spoti['album'] = df_spoti['album'].apply(minusculas)
```

```python
'''Sacamos cinco filas aleatorias'''

df_spoti['title'].sample(5)
```


- Algunos de los artistas tienen sÃ­mbolos raros, en concreto el sÃ­mbolo $. Cread una funciÃ³n que nos elimine este sÃ­mbolo de los nombres de los artistas.

```python
df_spoti.columns
```

```python
'''Con un for loop comprobamos si tenemos el sÃ­mbolo del dolar en "artist" '''

for dolar in df_spoti['artist']:
    if '$' in dolar:
        print(dolar)
```

```python
'''Hacemos la funciÃ³n donde le pedimos que en caso de que el nombre sea de tipo string, me reemplace el simbolo por una 's' '''
def dolar(nombre):
    if isinstance(nombre, str):
        return nombre.replace("$", "S")
    else:
        return nombre
```

```python
''' Hacemos el apply de nuestra funciÃ³n'''

df_spoti['artist'] = df_spoti['artist'].apply(dolar)
```

Ahora queremos analizar la colaboraciÃ³n entre artistas en la industria musical, se tienen datos de las colaboraciones realizadas en los Ãºltimos aÃ±os. Se desea conocer cuÃ¡les son los artistas que han colaborado en mayor cantidad de canciones y generar un ranking con los diez primeros. Â¿QuÃ© artistas con los que mÃ¡s colaboraciones tienen? Para esto tendrÃ©is que:

- Filtrad los datos para quedaros solo con aquellos datos donde en la columna "track" tengamos "feat".

```python
'''Establecemos el patrÃ³n que queremos para despues aplicarlo al mÃ©todo str.contains()'''

patron = 'feat'

df_feat = df_spoti[df_spoti['track'].str.contains(patron, regex= True)]

df_feat.sample(5)
```

- Contad cuantas canciones tiene cada artista utilizando el dataframe generado en el paso anterior.

```python
'''He hecho un .count() con el gropuby para que me devuelva las observaciones no nulas'''

df_numero_feat = df_feat.groupby('artist')['title'].count().reset_index()

df_numero_feat
```


- Mostrad los resultados en una grÃ¡fica, seleccionado los 10 artistas con mÃ¡s colaboraciones.

```python
'''Hemos hecho un sort_values del dataframe anterior con ascending false para que las ordene de mayor a menor 
y hemos sacado los diez primeros artistas por tÃ­tulo'''

df_top_feat = df_numero_feat.sort_values('title', ascending= False).head(10)
df_top_feat
```

```python
plt.figure(figsize = (7,7))

fig = sns.barplot(x = 'title', y = 'artist', data = df_top_feat, palette = 'icefire')

plt.show()
```
